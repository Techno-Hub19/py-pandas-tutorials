```python
import numpy as np
import pandas as pd
```

```python
subs = pd.read_csv('subs.csv')
```

```python
kohli = pd.read_csv('kohli_ipl.csv')
```

```python
bollywood = pd.read_csv('bollywood.csv')    
```

```python
# astype
import sys
sys.getsizeof(kohli)
```

```python
# between
kohli[kohli['match_no'].between(1,15)]
```

```python
# clip
subs.clip(lower=100, upper=500).tail()
```

```python
# drop_duplicates
temp = pd.Series([1,2,2,3,4,4,5,5,5,6])
temp.drop_duplicates()
```

```python
bollywood.drop_duplicates(subset=[bollywood.columns[1]])
```

```python
# duplicated
temp = pd.Series([1,2,2,3,4,4,5,5,5,6])
temp.duplicated()   # show boolean values
```

```python
kohli.duplicated().sum()  # count duplicates
```

```python
temp = pd.Series([1,2,np.nan,3,np.nan,4,5])
temp.count()  # count non-null values
```

```python
temp.size # total size
```

```python
# is null
temp.isnull()  # check for null values`
```

```python
# dropna
temp.dropna()  # drop null values
```

```python
# fillna
temp.fillna(temp.mean())  # fill null values with mean
```

```python
#isin
temp.isin([2,4,6])  # check if values are in the list 
```

```python
kohli['runs'].isin([0,100]) # check if runs are 0 or 100
```

```python
# apply - apply a function to each element & custom functions
bollywood.apply(lambda x: x.str.split().str[0])  # get first word of each title
```
