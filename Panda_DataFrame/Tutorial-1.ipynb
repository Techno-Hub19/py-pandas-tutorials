# Creating DataFrame 

```python
# using list

student_data = [
    [100,70,45],
    [90,88,95],
    [80,75,85],
    [70,65,75]
]

student_data = pd.DataFrame(student_data, columns=['IQ','Marks','Package'])
student_data
```

```python
# using dict 

studentData = {
    'IQ':[100,90,80,70],
    'Marks':[70,88,75,65],
    'Package':[45,95,85,75]
}

pd.DataFrame(studentData)
```
