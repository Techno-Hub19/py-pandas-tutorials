# Boolean Indexing on series

```python
# find no's of 50's and 100's scored by kohli
ipl[ipl >= 50].size
```

```python
# find no of ducks
ipl[ipl==0].size
```

```python
# Count a number of day when i had more than 200 subs
subs[subs>200].size
```

```python
# find actors who have done more than 20 movies
num_movies = movie.value_counts()
num_movies[num_movies > 20]
```
