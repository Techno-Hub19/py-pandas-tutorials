## Pandas
- Pandas is a fast,powerful, flexible and easy to use open source data analysis and manipulation tool, built on the top of python programming language 

## Pandas Series
- A pandas series is like a column of table. It is 1-D array holding data of any type

```python
# import pandas library
import numpy as np
import pandas as pd
```

## Series from lists

```python
# String
country = ['India','USA','Brazil','New zealand','Australia']
pd.Series(country)
```

```python
# Integers
runs = [12,34,54,6,67,23]
pd.Series(runs)
```

```python
# Custom Index

marks = [35,65,87,34,87]
subject = ['English','Maths','History','Geography','Marathi']

pd.Series(marks,subject)
```

```python
# Setting a Name

marks = [35,65,87,34,87]
subject = ['English','Maths','History','Geography','Marathi']

pd.Series(marks,index=subject,name ="Subject-wise Marks")
```

# Series from Dict

```python
marks = {
    'Maths':85,
    'English':50,
    'Marathi':67,
    'History':56,
    'Geography':87
    }

marks_subject = pd.Series(marks,name = "Marks with Dict")
marks_subject
```

# Series Attribute

```python
# size
marks_subject.size
```

```python
# dtype
marks_subject.dtype
```

```python
# name
marks_subject.name
```

```python
# isUnique
pd.Series([1,2,3,1,4,5]).is_unique
```

```python
# index
marks_subject.index
```

```python
pd.Series(runs).index
```

```python
# values
pd.Series(country).values
```

```python
pd.Series(marks_subject).values
```
