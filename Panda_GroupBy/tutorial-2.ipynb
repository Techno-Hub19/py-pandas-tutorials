```python
# import lib 
import numpy as np
import pandas as pd
```

```python
# importing files

movies = pd.read_csv('imdb-top-1000.csv')
delivery = pd.read_csv('deliveries.csv')
```

# Aggreagation Methods

```python
genre = movies.groupby("Genre")
```

```python
# aggregation functions
# passing dict --> can apply different agg func at same-time on group-by

genre.agg(
    {
        'Runtime':'mean',
        'IMDB_Rating':'mean',
        'No_of_Votes':'sum',
        'Metascore':'min'
    }
)
```

```python
# passing list

genre.agg(['max','min'])
```

```python
# specific agg func on any column
genre.agg(
    {
        'Runtime': ['mean','min'],
        'IMDB_Rating':'mean',
        'No_of_Votes':['sum','max'],
        'Metascore':'min'
    }
)
```

```python
# looping on groups
for group,data in genre:
    print(data)
```

```python
# split apply combine
# apply - builtin function --> custom logic 
genre.apply(min)
```

```python
# find number of movies starting with A for each group
def Starts_A(group):
    return group['Series_Title'].str.startswith('A').sum()
```

```python
genre.apply(Starts_A)   # split-apply-combine
```

```python
# find ranking of each movie in the group according to IMDB rating
def rank(group):
    group['Genre_Rank'] = group['IMDB_Rating'].rank(ascending = False)
    return group
```

```python
genre.apply(rank)
```

```python
# find normalize IMDB rating group wise
def normalization(group):
    group['Norm_Rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min()) / (group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
    return group
```

```python
genre.apply(normalization)
```

```python
# find the top 10 batsman in terms of runs scored

batsmen = delivery.groupby('batsman')
batsmen['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# find the batsman with max no of sixes
most_sixes = delivery[delivery['batsman_runs'] == 6]
most_sixes.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1)
```

```python
# find batsman with most number of 4's and 6's in last 5 over
td = delivery[delivery['over'] > 15]
td = td[(td['batsman_runs'] == 4) | (td['batsman_runs'] == 6)]
```

```python
td.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1)
```

```python
# create a function that can return the highest score of any batsman
def highest_runs(batsmen):
    temp = delivery[delivery['batsman'] == batsmen]
    high_runs = temp.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
    return high_runs
```
