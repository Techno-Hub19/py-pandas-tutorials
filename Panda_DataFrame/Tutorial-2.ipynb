```python
# using dict 

studentData = {
    'name':['Shubham','sakshi','vinaya','Mogambo'],
    'IQ':[100,90,80,70],
    'Marks':[70,88,75,65],
    'Package':[45,95,85,75]
}

studentData = pd.DataFrame(studentData)

# set_index --> to set a particular column as index
studentData.set_index('name', inplace=True) # inplace makes the changes permanent
studentData
```

# Selecting rows from a DataFrame using loc and iloc
- <b>iloc</b> : Searches using index positions 
- <b>loc</b> : Searches using index labels

```python
# iloc --> index location based indexing
# single row operations

movies.iloc[1] # iloc give a single row 
#dtype --> Series
```

```python
# multiple row operations

movies.iloc[0:5] # dataFrame
```

```python
# fancy indexing --> fancy selection of rows
movies.iloc[[0,5,10,15,20]]  # DataFrame
```

```python
# loc --> label based indexing

studentData.loc['Shubham']
```

```python
studentData.loc['Shubham':'vinaya']  # DataFrame
```

```python
studentData.loc[['Shubham','Mogambo','vinaya']]  # fancy indexing
```

```python
student_data.iloc[[0,3]]     # Fancy indexing rows 0 and 3
```

# Selecting Both rows and columns 

```python

student_data.iloc[0:3, 1:3]  # rows 0 to 2 and columns 1 to 2
```

```python
movies.iloc[0:3,0:3]
```

```python
movies.loc[0:3, ['title_x', 'imdb_id', 'poster_path']]
```
