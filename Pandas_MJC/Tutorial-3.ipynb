# IPL Problem 


```python
# find top 3 stadiums with highest sixes/match ratio

temp_df = delivery.merge(match,left_on="match_id",right_on="id")
temp_df
```

```python
six_df = temp_df[temp_df["batsman_runs"] == 6]
six_df
```

```python
# sixes on each stadium
num_sixes = six_df.groupby("venue")['venue'].count()
num_sixes
```

```python
num_matches = match['venue'].value_counts()
num_matches
```

```python
(num_sixes/num_matches).sort_values(ascending=False).head(3)
```

```python
# Find orange cap holder for each season
temp_df = delivery.merge(match,left_on="match_id",right_on="id")
temp_df.groupby(["season","batsman"])["batsman_runs"].sum().reset_index().sort_values(by="batsman_runs",ascending=False).drop_duplicates(subset="season",keep="first").sort_values(by="season")
```
