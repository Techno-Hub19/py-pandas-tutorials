```python
# importing libraries
import numpy as np
import pandas as pd
```

# Concatenation

```python
# importing files
courses = pd.read_csv("courses.csv")
delivery = pd.read_csv("deliveries.csv")
nov = pd.read_csv("reg-month1.csv")
dec = pd.read_csv("reg-month2.csv")
students = pd.read_csv("students.csv")
```

```python
# pd.concat --> default merges vertically --> but indexes are repeated/retan
pd.concat([nov,dec])
```

```python
pd.concat([nov,dec]).shape
```

```python
# ignore_index
regs = pd.concat([nov,dec],ignore_index=True)  # it will ignore repeating of index
regs
```

```python
# multi-index DataFrame
multi = pd.concat([nov,dec],keys=["nov","dec"]) # keys for multi-indexing
multi.loc['nov']    # only get "nov" based data from Df 
```

```python
multi.loc[("nov",0)]    # 1st person from "nov" index
```

```python
# concate vertically --> doesn't matter shape of data
pd.concat([nov,dec],axis=1) # axis = horizontal(column)
```

```python
pd.concat([nov,dec],axis=1).shape
```
