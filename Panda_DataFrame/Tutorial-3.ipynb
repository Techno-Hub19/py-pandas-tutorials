```python
# Write a function that can return the track reord of 2 teams against each other.

def match_record(team1, team2):
    matches = ipl[((ipl['Team1'] == team1) & (ipl['Team2'] == team2)) | ((ipl['Team1'] == team2) & (ipl['Team2'] == team1))]
    team1_wins = matches[matches['WinningTeam'] == team1].shape[0]
    team2_wins = matches[matches['WinningTeam'] == team2].shape[0]
    return f"{team1} has won {team1_wins} matches, {team2} has won {team2_wins} matches out of {matches.shape[0]} matches."     
```

```python
match_record('Chennai Super Kings', 'Mumbai Indians')
```
