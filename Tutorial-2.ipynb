# Editing Series

```python
marks = {
    'Maths':85,
    'English':50,
    'Marathi':67,
    'History':56,
    'Geography':87
    }

marks_subject = pd.Series(marks,name = "Marks with Dict")
marks_subject
```

```python
# Using indexing 
marks_subject[1] = 100
marks_subject
```

```python
# What if an index does not exist
marks_subject['EVS'] = 76
marks_subject
```

```python
# Slicing
runs = ipl.head(5)
runs
```

```python
runs[1:3] = [50,24]
runs
```

```python
# Fancy Indexing
runs[[1,3,5]] = [20,40,59]
runs
```

```python
# using label indexing
movie['Saathiya (film)'] = 'Ajay Devgan'
movie
```
